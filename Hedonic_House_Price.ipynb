```python
# 5.3 Python Code for Model 1
# The following code implements the regression for RQ1:

#Python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
# Ideally, the file path would be dynamic, but here we reference the uploaded file
try:
    df = pd.read_csv('Lab11-RealEstate.csv')

    # Rename columns for easier access based on common dataset structure and usage in code
    df.columns = [
        'No',
        'transaction_date',
        'House_Age',
        'Distance_to_MRT',
        'num_convenience_stores',
        'latitude',
        'longitude',
        'Price'
    ]

    # --- Research Question 1: Price vs Distance to MRT ---
    X1 = df[['Distance_to_MRT']]
    y = df[['Price']]

    # Split data into training and testing sets (80% train, 20% test)
    X_train1, X_test1, y_train1, y_test1 = train_test_split(
        X1, y, test_size=0.2, random_state=42
    )

    # Initialize and train the model
    model_mrt = LinearRegression()
    model_mrt.fit(X_train1, y_train1)

    # Make predictions
    y_pred1 = model_mrt.predict(X_test1)

    # Output results
    print("--- Model 1 Results (Distance to MRT) ---")
    print(f"Intercept: {model_mrt.intercept_[0]:.4f}")
    print(f"Slope (Coefficient): {model_mrt.coef_[0][0]:.5f}")
    print(f"Mean Squared Error: {mean_squared_error(y_test1, y_pred1):.4f}")
    print(f"R-squared: {r2_score(y_test1, y_pred1):.4f}")

    # Visualization
    plt.figure(figsize=(10, 6))
    plt.scatter(X_test1, y_test1, color='blue', alpha=0.6, label='Actual Data')
    plt.plot(X_test1, y_pred1, color='red', linewidth=2, label='Linear Fit')
    plt.title('Relationship between House Price and Distance to MRT')
    plt.xlabel('Distance to Nearest MRT Station (m)')
    plt.ylabel('House Price (10k NTD/Ping)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # 6.3 Python Code for Model 2
    # The following code implements the regression for RQ2:
    # Python
    # --- Research Question 2: Price vs House Age ---
    X2 = df[['House_Age']]
    # y is already defined

    # Split data (using the same random state for comparability)
    X_train2, X_test2, y_train2, y_test2 = train_test_split(
        X2, y, test_size=0.2, random_state=42
    )

    # Initialize and train the model
    model_age = LinearRegression()
    model_age.fit(X_train2, y_train2)

    # Make predictions
    y_pred2 = model_age.predict(X_test2)

    # Output results
    print("\n--- Model 2 Results (House Age) ---")
    print(f"Intercept: {model_age.intercept_[0]:.4f}")
    print(f"Slope (Coefficient): {model_age.coef_[0][0]:.5f}")
    print(f"Mean Squared Error: {mean_squared_error(y_test2, y_pred2):.4f}")
    print(f"R-squared: {r2_score(y_test2, y_pred2):.4f}")

    # Visualization
    plt.figure(figsize=(10, 6))
    plt.scatter(X_test2, y_test2, color='green', alpha=0.6, label='Actual Data')
    plt.plot(X_test2, y_pred2, color='orange', linewidth=2, label='Linear Fit')
    plt.title('Relationship between House Price and House Age')
    plt.xlabel('House Age (years)')
    plt.ylabel('House Price (10k NTD/Ping)')
    plt.legend()
    plt.grid(True)
    plt.show()

except FileNotFoundError:
    print("Error: The file 'Lab11-RealEstate.csv' was not found.")
```
